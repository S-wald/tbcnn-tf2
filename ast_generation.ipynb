```python
import tensorflow as tf
children_vectors = tf.constant([[[1.0], [2.0], [3.0]], [[2.0], [2.0], [3.0]]], dtype=tf.float64)
```

```python
children_vectors
```

```python
num_children = tf.shape(children_vectors)[-2]
```

```python
num_children
```

```python
children_indices = tf.range(1, num_children+1, 1)
```

```python
children_indices
```

```python
eta_l = tf.where(
    tf.equal(num_children, 1),
    tf.constant(1, dtype=tf.float64),
    tf.divide(num_children - children_indices, num_children - 1)
)
```

```python
eta_r = tf.where(
    tf.equal(num_children, 1),
    tf.constant(1, dtype=tf.float64),
    tf.divide(children_indices - 1, num_children - 1)
)
```

```python
eta_l
```

```python
eta_r
```

```python
batch_size = tf.shape(children_vectors)[0]
```

```python
eta_l_batch = tf.tile(tf.expand_dims(eta_l, axis=1), [batch_size, 1])
```

```python
eta_r_batch = tf.tile(tf.expand_dims(eta_r, axis=1), [batch_size, 1])
```

```python
eta_l_batch
```

```python
eta_r_batch
```

```python
coef = tf.stack([eta_l_batch, eta_r_batch], axis=2)
```

```python
coef
```

```python
coef_reshape = tf.reshape(coef, (x, tf.cast(y, dtype=tf.int32), 2))
```
