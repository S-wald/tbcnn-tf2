```python
import os
import re
path = r"C:\Users\sebas\Documents\Masterarbeit\datasets\1_clean_datasets\cpp14_100_1000"

fname = []
for root,d_names,f_names in os.walk(path):
    for f in f_names:
        fname.append(os.path.join(root, f))
with open(os.path.join(path, "ai4code_spt_input.txt"), "w") as f:
    for fn in fname:
        if "ai4code" in fn:
            continue
        problem_id = re.findall(r"p\d{5}", fn)[0]
        f.write(fn + " " + path + f"\\ai4code_spt\\{problem_id}\\\n")
```
