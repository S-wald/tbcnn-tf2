```python
samples_per_language = 1000
```

```python
from dataset.CodeNet import CodeNet
code_net = CodeNet(location=r"C:\Users\sebas\Documents\Masterarbeit\Project_CodeNet")
```

```python
submission_metadata = code_net.get_submission_metadata()
submission_metadata.info()
```

```python
submission_metadata.groupby(["language"]).size().reset_index(name="count").sort_values("count", ascending=False)
```

```python
accepted_cpp_submissions = submission_metadata.query("language == 'C++' and status == 'Accepted'")
accepted_cpp_submissions.groupby(["original_language"]).size().reset_index(name="count").sort_values("count", ascending=False)
```

```python
accepted_cpp14_submissions = accepted_cpp_submissions.query("original_language in ['C++14 (GCC 5.4.1)', 'C++14', 'C++14 (Clang 3.8.0)']")
accepted_cpp14_submissions
```

```python
cpp14_submissions_count = accepted_cpp14_submissions.groupby(["problem_id"]).size().reset_index(name="count").sort_values("count", ascending=False).query(f"count >= {samples_per_language}")
cpp14_submissions_count
```

```python
accepted_python_submissions = submission_metadata.query("language == 'Python' and status == 'Accepted'")
accepted_python_submissions.groupby(["original_language"]).size().reset_index(name="count").sort_values("count", ascending=False)
```

```python
accepted_python3_submissions = accepted_python_submissions.query("original_language in ['Python (3.4.3)', 'Python (3.8.2)', 'Python (3.4.2)']")
accepted_python3_submissions
```

```python
python3_submissions_count = accepted_python3_submissions.groupby(["problem_id"]).size().reset_index(name="count").sort_values("count", ascending=False).query(f"count >= {samples_per_language}")
python3_submissions_count
```

```python
accepted_java_submissions = submission_metadata.query("language == 'Java' and status == 'Accepted'")
accepted_java_submissions.groupby(["original_language"]).size().reset_index(name="count").sort_values("count", ascending=False)
```

```python
accepted_java8_submissions = accepted_java_submissions.query("original_language in ['Java8 (OpenJDK 1.8.0)', 'Java (OpenJDK 1.8.0)']")
accepted_java8_submissions
```

```python
java8_submissions_count = accepted_java8_submissions.groupby(["problem_id"]).size().reset_index(name="count").sort_values("count", ascending=False).query(f"count >= {1}")
java8_submissions_count
```

```python
merged_cpp_python = cpp14_submissions_count.merge(python3_submissions_count, on="problem_id")
merged_cpp_python
```

```python
merged_cpp_python_java = merged_cpp_python.merge(java8_submissions_count, on="problem_id")
merged_cpp_python_java
```

```python
problem_ids = merged_cpp_python_java['problem_id'].unique()
problem_ids
```

```python
import os
import shutil
target_dir = r"C:\Users\sebas\Documents\Masterarbeit\datasets\cpp14_python3_java8_400"
for problem_id in problem_ids:
    cpp14_samples = accepted_cpp14_submissions.query(f'problem_id == "{problem_id}"').sample(samples_per_language)
    python3_samples = accepted_python3_submissions.query(f'problem_id == "{problem_id}"').sample(samples_per_language)
    java8_samples = accepted_java8_submissions.query(f'problem_id == "{problem_id}"').sample(samples_per_language)

    problem_id_trg_dir = os.path.join(target_dir, problem_id)
    os.mkdir(problem_id_trg_dir)

    problem_id_cpp14_trg_dir = os.path.join(problem_id_trg_dir, 'cpp14')
    os.mkdir(problem_id_cpp14_trg_dir)
    submission_ids = cpp14_samples["submission_id"].values
    paths = code_net.get_src_paths_of_submissions(list(submission_ids))
    for _, value in paths.items():
        shutil.copy(value, problem_id_cpp14_trg_dir)

    problem_id_python3_trg_dir = os.path.join(problem_id_trg_dir, 'python3')
    os.mkdir(problem_id_python3_trg_dir)
    submission_ids = python3_samples["submission_id"].values
    paths = code_net.get_src_paths_of_submissions(list(submission_ids))
    for _, value in paths.items():
        shutil.copy(value, problem_id_python3_trg_dir)

    problem_id_java8_trg_dir = os.path.join(problem_id_trg_dir, 'java8')
    os.mkdir(problem_id_java8_trg_dir)
    submission_ids = java8_samples["submission_id"].values
    paths = code_net.get_src_paths_of_submissions(list(submission_ids))
    for _, value in paths.items():
        shutil.copy(value, problem_id_java8_trg_dir)

```
